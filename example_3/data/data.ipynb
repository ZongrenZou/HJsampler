```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

from scipy.integrate import odeint
```

```python
def solve(y0, t, eps=0.01):
    ys = [y0]
    dt = t[1] - t[0]
    diffs = []
    print(dt)
    print(t.shape, dt)
    for i in range(t.shape[0] - 1):
        ys += [ys[-1] + (np.sin(4*np.pi*t[i, 0]) + 1 * ys[-1]*ys[-1]) * dt + np.sqrt(eps * dt) * np.random.normal(size=y0.shape)]
    return ys
```

```python
T = 1
y0 = 0. + 0.1 * np.random.normal(size=[1, 1000000])

t = np.linspace(0, 1, 101).reshape([-1, 1])
eps = 0.0005
ys = solve(y0, t, eps=eps)
print(ys[-1].shape)
ys = np.stack(ys, axis=0)[:, 0, :].T
```

```python
sio.savemat(
    "./train5.mat",
    {
        "t": t.reshape([-1, 1]),
        "y": ys,
        "eps": eps
    }
)
```
