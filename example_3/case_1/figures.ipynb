```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
```

#### Case 1

```python
data = sio.loadmat("./outputs/sgm1.mat")
t = data["t"]
mu = data["mu"].flatten()
sd = data["sd"].flatten()
sols = data["sols"]
```

```python
fig = plt.figure(dpi=100)
plt.plot(t.flatten(), sols, "k-", label="Reference", linewidth=2)

plt.plot(t.flatten(), mu, "r--", label="Mean", linewidth=2)
plt.fill_between(
    t.flatten(), 
    mu + 2*sd, mu - 2*sd, 
    alpha=0.5, 
    label="2 SD",
    color="c",
    edgecolor="none",
)
plt.ylim([-0, 0.5])
plt.xlabel("$t$")
plt.legend()
plt.title("$\epsilon=0.001$")
plt.savefig("./figs/sgm_1.png")
plt.show()
```

#### Case 0

```python
data = sio.loadmat("./outputs/sgm1.mat")
t = data["t"]
sols = data["sols"]
z_sols = data["z_sols"]
```

```python
fig = plt.figure(dpi=100)
plt.plot(t.flatten(), sols, "k-", label="Reference", linewidth=2)
plt.plot(t.flatten(), np.flip(z_sols, axis=0), "r--", label="Inference", linewidth=2)

plt.ylim([-0.0, 0.5])
plt.xlabel("$t$")
plt.legend()
plt.title("Consequence of model misspecification")
plt.savefig("./figs/sgm_0.png")
plt.show()
```
