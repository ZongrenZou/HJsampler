```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
```

#### Figure 1

```python
data = sio.loadmat("./outputs/case_1.mat")
riccati = data["samples_riccati"]
sgm = data["samples_sgm"]
```

```python
plt.figure(dpi=100)
plt.hist2d(riccati[:, 0], riccati[:, 1], bins=(100, 100), density=True, cmap=colormaps["jet"])
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])

# plt.axis("off")
plt.title("$Y_1 = [0.5, -0.5]^T$", fontsize=20)
plt.plot([0.5], [-0.5], "wo", markersize=10)
plt.clim([0, 0.8])
plt.colorbar()

plt.tight_layout()

plt.savefig("./figs/riccati_1.png")
plt.show()
```

```python
plt.figure(dpi=100)
plt.hist2d(sgm[:, 0], sgm[:, 1], bins=(100, 100), density=True, cmap=colormaps["jet"])
plt.xlim([-1.5, 1.5])
plt.ylim([-1.5, 1.5])

# plt.axis("off")
plt.title("$Y_1 = [0.5, -0.5]^T$", fontsize=20)
plt.plot([0.5], [-0.5], "wo", markersize=10)
plt.clim([0, 0.8])
plt.colorbar()

plt.tight_layout()

plt.savefig("./figs/sgm_1.png")
plt.show()
```
