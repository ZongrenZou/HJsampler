```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
sigma = 1
eps = 1
yt = 2
T = 1
dt = 0.0001
B = 2
```

```python
Ps = [sigma ** 2 / eps]

for i in range(int(T/dt)):
    update = 1 - Ps[-1] * B - B * Ps[-1]
    Ps += [Ps[-1] + dt * update]

Ps = np.array(Ps)
```

```python
t = np.linspace(0, 1, int(T/dt) + 1)
Ps_ref = np.exp(-2*B*t) * sigma ** 2 / eps + (1 - np.exp(-2*B*t)) / 2 / B
```

```python
plt.plot(t, Ps_ref)
plt.plot(t, Ps)
```

```python
np.mean((Ps - Ps_ref) ** 2)
```
