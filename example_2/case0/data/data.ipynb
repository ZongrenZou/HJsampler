```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
from scipy.linalg import expm
```

```python
alpha = 3
eps = 1.5
T = 1
y0 = np.random.normal(size=[1000000, 1])
```

```python
Nt = 101
t = np.linspace(0, T, Nt)
ys = [y0]
dt = t[1] - t[0]
for i in range(int(T/dt)):
    b = -alpha * ys[-1]
    ys += [ys[-1] + b * dt + np.sqrt(dt * eps) * np.random.normal(size=y0.shape)]
ys = np.stack(ys)
ys = np.transpose(ys, [1, 0, 2])
```

```python
sio.savemat(
    "./train.mat",
    {
        "t": t,
        "ys": ys,
        "eps": eps,
        "alpha": alpha,
        "T": T
    }
)
```
