```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

from scipy.integrate import odeint
```

```python
def rhs(y, t):
    y1, y2 = y
    dydt = [y2, -10*y1 - 1*y2]
    return dydt
```

```python
y0 = [0, 0.1]
t = np.linspace(0, 1, 1001)
sols = odeint(rhs, y0, t)
```

```python
plt.plot(t, sols[:, 0], "k-")
plt.plot(t, sols[:, 1], "r-")
plt.show()
```

```python
sols[-1, :]
```

```python
def rhs(y, t):
    y1, y2 = y
    dydt = [-y2, +10 * y1 + 1 * y2]
    return dydt
```

```python
z0 = sols[-1, :]
t = np.linspace(0, 1, 1001)
z_sols = odeint(rhs, z0, t)
```

```python
plt.plot(t, sols[:, 0], "k-")
plt.plot(t, sols[:, 1], "r-")
plt.plot(t, np.flip(z_sols[:, 0]), "b--")
plt.plot(t, np.flip(z_sols[:, 1]), "b--")
plt.show()
```
