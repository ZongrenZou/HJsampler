```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
```

#### Case 1

```python
data = sio.loadmat("./outputs/sgm1.mat")
t = data["t"]
mu = data["mu"]
sd = data["sd"]
sols = data["sols"]
z_sols = data["z_sols"]
```

```python
fig = plt.figure(dpi=200)
plt.plot(t.flatten(), sols[:, 0], "k-", label="Reference of $y_1$", linewidth=2)
plt.plot(t.flatten(), sols[:, 1], "b-", label="Reference of $y_2$", linewidth=2)

plt.plot(t.flatten(), np.flip(mu[:, 0]), "r--", label="Mean of inferred $y_1$")
plt.fill_between(
    t.flatten(), 
    np.flip(mu[:, 0] + 2*sd[:, 0]), 
    np.flip(mu[:, 0] - 2*sd[:, 0]), 
    alpha=0.3, 
    label="2 SD of inferred $y_1$", 
    color="r",
    edgecolor="none",
)
plt.plot(t.flatten(), np.flip(mu[:, 1]), "g--", label="Mean of inferred $y_2$")
plt.fill_between(
    t.flatten(), 
    np.flip(mu[:, 1] + 2*sd[:, 1]), 
    np.flip(mu[:, 1] - 2*sd[:, 1]), 
    alpha=0.3, 
    label="2 SD of inferred $y_2$", 
    color="b",
    edgecolor="none",
)
plt.ylim([-0.2, 0.4])
plt.xlabel("$t$")
plt.legend()
plt.title("$\epsilon=1\\times 10^{-4}$")
plt.savefig("./figs/sgm_1.png")
plt.show()
```
