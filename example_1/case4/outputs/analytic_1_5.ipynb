```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
```

```python
fig, axes = plt.subplots(1, 5, figsize=[15, 3], dpi=500)

data1 = sio.loadmat("./analytic_1.mat")
x = data1["x"].flatten()
y = data1["y"].flatten()
samples = data1["zt"].flatten()

axes[0].hist(samples, density=True, color="c", bins=100)
axes[0].plot(x, y, "k--", label="The exact posterior", linewidth=1)
axes[0].set_title("$P(Y_0|Y_1=-0.3)$")


data2 = sio.loadmat("./analytic_2.mat")
y = data2["y"].flatten()
samples = data2["zt"].flatten()

axes[1].hist(samples, density=True, color="c", bins=500)
axes[1].plot(x, y, "k--", label="The exact posterior", linewidth=1)
axes[1].set_title("$P(Y_0|Y_{0.3}=-0.2)$")


data3 = sio.loadmat("./analytic_3.mat")
y = data3["y"].flatten()
samples = data3["zt"].flatten()

axes[2].hist(samples, density=True, color="c", bins=500)
axes[2].plot(x, y, "k--", label="The exact posterior", linewidth=1)
axes[2].set_title("$P(Y_0|Y_{0.5}=-0.1)$")


data4 = sio.loadmat("./analytic_4.mat")
y = data4["y"].flatten()
samples = data4["zt"].flatten()

axes[3].hist(samples, density=True, color="c", bins=500)
axes[3].plot(x, y, "k--", label="The exact posterior", linewidth=1)
axes[3].set_title("$P(Y_0|Y_{0.8}=0.0)$")


data5 = sio.loadmat("./analytic_5.mat")
y = data5["y"].flatten()
samples = data5["zt"].flatten()

axes[4].hist(samples, density=True, color="c", bins=500)
axes[4].plot(x, y, "k--", label="The exact posterior", linewidth=1)
axes[4].set_title("$P(Y_0|Y_{0.7}=0.2)$")


for i in range(5):
    # axes[i].set_ylim([-0.05, 0.55])
    axes[i].set_xlim([-0.9, 0.9])
    pass

fig.savefig("./analytic.png")
plt.show()
```
