```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import scipy.io as sio
import tensorflow as tf
from tqdm import trange


import models
```

#### Case 1

```python
data1 = sio.loadmat("./outputs/sgm1.mat")
xx = data1["xx"]
yy = data1["yy"]
f = data1["f"]

plt.figure(dpi=100)
plt.pcolormesh(
    xx.reshape([1001, 1001]), 
    yy.reshape([1001, 1001]),
    f,
    cmap=colormaps["jet"],
    vmin=0,
    vmax=0.85,
)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.colorbar()

# plt.axis("off")
plt.title("$P(Y_{0.1}| Y_{0.9}= [-0.9, 0.9]^T)$", fontsize=20)
plt.plot([-0.9], [0.9], "wo", markersize=10)
plt.tight_layout()

plt.savefig("./figs/ref1.png")
plt.show()
```

```python
data1 = sio.loadmat("./outputs/sgm1.mat")
samples = data1["samples"]

plt.figure(dpi=100)
plt.hist2d(samples[:, 0], samples[:, 1], bins=(100, 100), density=True, cmap=colormaps["jet"])
plt.xlim([-1, 1])
plt.ylim([-1, 1])

# plt.axis("off")
plt.title("$P(Y_{0.1}| Y_{0.9}= [-0.9, 0.9]^T)$", fontsize=20)
plt.plot([-0.9], [0.9], "wo", markersize=10)
plt.clim([0, 0.85])
plt.colorbar()

plt.tight_layout()

plt.savefig("./figs/sgm1.png")
plt.show()
```
