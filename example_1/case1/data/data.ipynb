```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
sigma1 = 0.5
sigma2 = 0.8
sigma3 = 0.6
mu1 = 0
mu2 = 2
mu3 = -2

eps = 5
T = 1

y01 = mu1 + sigma1 * np.random.normal(size=[1000000, 1])
y02 = mu2 + sigma2 * np.random.normal(size=[1000000, 1])
y03 = mu3 + sigma3 * np.random.normal(size=[1000000, 1])
y0 = np.concatenate(
    [y01, y02, y03], axis=0,
)
```

```python
Nt = 101
t = np.linspace(0, T, Nt)
ys = [y0]
dt = t[1] - t[0]
for i in range(int(T/dt)):
    ys += [ys[-1] + np.sqrt(dt * eps) * np.random.normal(size=y0.shape)]
ys = np.stack(ys)[..., 0].T
```

```python
sio.savemat(
    "./train.mat",
    {
        "t": t,
        "ys": ys,
        "eps": eps,
        "T": T,
    }
)
```
